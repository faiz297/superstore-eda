# Data Analysis

### Data Quality Check

```python
##checking for missing values
missing_data=df.isnull().sum()
print(missing_values[missing_data > 0]) #filters only columns where missing values exist
print("Duplicate rows: ",df.duplicated().sum())
```

## Data Type Conversion

```python
#Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
print("New Order Date",df['Order Date'].dtype)
print("New Ship Date",df['Ship Date'].dtype)
```

## Univariate Analysis

```python

print("\n ===CATEGORICAL VARIABLES=== \n")

categorical_cols=['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']
for cols in categorical_cols:
    print(f'--- {cols}---')
    print(df[cols].value_counts())
    #Visualization
    plt.figure(figsize=[4,4])
    df[cols].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f'Distribution of {cols}')
    plt.tight_layout()
    plt.xticks(rotation=30)
    plt.show()
```

## Numerical Variable Analysis

```python
print("\n ---NUMERICAL VARIABLE ANALYSIS--- \n")
numerical_cols=['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit','Stock Left','Reorder Quantity']
for cols in numerical_cols:
    print(f'---{cols}---')
    print(f"Mean: {df[cols].mean():.2f}")
    print(f"Median: {df[cols].median():.2f}")
    print(f"Standard Deviation: {df[cols].std():.2f}")
    print(f"Min: {df[cols].min():.2f}")
    print(f"Max: {df[cols].max():.2f}")
    #Visualization
    plt.figure(figsize=[8,4])
    plt.subplot(1,2,1)
    df[cols].hist(bins=30,color='lightgreen',alpha=0.7)
    plt.title(f'Distribution of {cols}')
    plt.subplot(1,2,2)
    df.boxplot(column=cols)
    plt.title(f'Boxplot of {cols}')
    plt.tight_layout()
    plt.show()
```

## Bivariate Analysis

```python
#Bivariate Analysis on 2 or more columns

print("\n SALES AND PROFIT BY CATEGORY \n")
category_performance=df.groupby('Category').agg({
    'Sales Amount':'sum',
    'Profit':'sum',
    'Order ID':'count'
}).rename(columns={'Order ID':'Order Count'})#changing Order ID to Order count for better understanding and cleaner output
category_performance['Profit Margin']=(category_performance['Profit'] / category_performance['Sales Amount']) *100 #adding a new column profit margin
print(category_performance)

#Visualization
fig , axes = plt.subplots(2,2,figsize=(18,18)) #creates a grid of 2x2 total 4 plots .axes is the address of the plots

#sales by category
category_performance['Sales Amount'].plot(kind='bar',ax=axes[0,0],color='blue',alpha=0.7)
axes[0,0].set_title('Total Sales by Category')
axes[0,0].tick_params(axis='x',rotation=30)

#profit by category
category_performance['Profit'].plot(kind='bar',ax=axes[0,1],color='green',alpha=0.7)
axes[0,1].set_title('Total profit by category')
axes[0,1].tick_params(axis='x',rotation=30)

#profit margin
category_performance['Profit Margin'].plot(kind='bar',ax=axes[1,0],color='red',alpha=0.7)
axes[1,0].set_title('Profit Margin by Category')
axes[1,0].tick_params(axis='x',rotation=30)

#order count
category_performance['Order Count'].plot(kind='bar',ax=axes[1,1],color='yellow',alpha=0.7)
axes[1,1].set_title('Order Count by Category')
axes[1,1].tick_params(axis='x',rotation=30)
plt.tight_layout
plt.show()
```

## Regional Analysis

```python
print("\n REGIONAL ANALYSIS \n")
regional_performance=df.groupby('Region').agg({
    'Sales Amount':'sum',
    'Profit':'sum',
    'Order ID':'count'
}).rename(columns={'Order ID':'Order Count'})
print(regional_performance)
regional_performance['Profit Margin'] = (regional_performance['Profit'] / regional_performance['Sales Amount'] * 100)
print(regional_performance)
#visualization
plt.figure(figsize=[12,8])
plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind='pie',autopct='%1.1f%%',shadow=True)
plt.title('Sales distribution by region')

plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind='pie',autopct='%1.1f%%',shadow=True)
plt.title("Profit distribution by region")

plt.subplot(2,2,3)
sns.barplot(x=regional_performance.index,y='Order Count',data=regional_performance)
plt.title('Order Count by Region')
plt.xticks(rotation=30)

plt.subplot(2,2,4)
sns.barplot(x=regional_performance.index,y='Profit Margin',data=regional_performance)
plt.title('Profit Margin by Region')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()





```

## Time Series Analysis

```python
#Extract Time details
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y') #to convert from object datatype to date
df['Order Month']=df['Order Date'].dt.to_period('M') #breaking date into year-month format only
df['Order Year']=df['Order Date'].dt.year #breaking date into year format only
monthly_trends=df.groupby('Order Month').agg({
    'Sales Amount':'sum',
    'Profit':'sum',
    'Order ID':'count'
}).rename(columns={'Order ID':'Order Count'})
print(monthly_trends)

#visualization
plt.figure(figsize=[15,10])

plt.subplot(3,1,1)
monthly_trends['Sales Amount'].plot(kind='line',color='blue',marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales amount")
plt.grid(True) #turns grid line on plot


plt.subplot(3,1,2)
monthly_trends['Profit'].plot(kind='line',color='green',marker='o')
plt.title("Monthly Profit")
plt.ylabel("Profit")
plt.grid(True)


plt.subplot(3,1,3)
monthly_trends['Order Count'].plot(kind='line',color='black',marker='o')
plt.title("Monthly Order Count")
plt.grid(True)
plt.ylabel("Order Count")
plt.tight_layout()
plt.show()


```

## Advance Analysis

```python
print("\n CORRELATION ANALYSIS \n ")
numerical_df=df[['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit']]

corelation_matrix=numerical_df.corr()
print(corelation_matrix)
#heatmap
plt.figure(figsize=[10,8])
sns.heatmap(corelation_matrix,annot=True,cmap='coolwarm',center=0,square=True,linewidth=0.5)#annot shows value inside each box center means the middle value is 0
plt.tight_layout()
plt.show()

```

```python
print("\n CUSTOMER SEGMENT ANALYSIS \n")

segment_analysis=df.groupby(['Customer Segment']).agg({
    'Sales Amount':['sum','mean'],
    'Profit':['sum','mean'],
    'Order ID':'count',
    'Discount (%)':'mean'
}).round(2)
print(segment_analysis)

segment_analysis.columns=['Total Sales','Avg Sales','Total Profit','Avg Profit','Order Count','Avg Discount'] #renaming columns
print(segment_analysis)

#visualization
fig , axes = plt.subplots(2,2,figsize=(15,10))
segment_analysis['Total Sales'].plot(kind='bar',ax=axes[0,0],color='lightblue')
axes[0,0].set_title('Total Sales by customer segment')
axes[0,0].tick_params(axis='x',rotation=30)

segment_analysis['Total Profit'].plot(kind='bar',ax=axes[0,1],color='lightblue')
axes[0,1].set_title('Total Sales by customer segment')
axes[0,1].tick_params(axis='x',rotation=30)


segment_analysis['Avg Sales'].plot(kind='bar',ax=axes[1,0],color='lightblue')
axes[1,0].set_title('Average sales by customer segment')
axes[1,0].tick_params(axis='x',rotation=30)

segment_analysis['Avg Profit'].plot(kind='bar',ax=axes[1,1],color='lightblue')
axes[1,1].set_title('Average Profit by customer segment')
axes[1,1].tick_params(axis='x',rotation=30)
plt.tight_layout()
plt.show()


                                                      
    
```

```python
print("\n KEY INSIGHTS \n")

total_sales=df['Sales Amount'].sum()
total_profit=df['Profit'].sum()
total_orders=df['Order ID'].nunique()

avg_profit_margin=(total_profit / total_sales) * 100

print("---OVERALL PERFORMANCE---")
print(f'Total Sales are {total_sales:,.2f}')
print(f'Total Profit is {total_profit:,.2f}')
print(f'Total Orders are {total_orders:}')
print(f'Average Profit Margin {avg_profit_margin:,.2f}')

top_categories=df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("TOP CATEGORIES ARE")
print(top_categories)

for i, (category, profit) in enumerate(top_categories.items(), 1): print(f"(1). (category): $[{profit:,.2f}")
#Regional Performance
best_region=df.groupby('Region') ['Profit'].sum().idxmax()
print("\nBEST PERFORMING REGION: (best_region)")
#Delivery Performance
delivery_stats=df[ 'Delivery Status'].value_counts(normalize=True) *100
print("\nDELIVERY STATUS: ")
for status, percentage in delivery_stats.items(): print(f"(status): {percentage:.1f}%")
#Customer Segments
best_segment = df.groupby('Customer Segment')["Profit"].sum().idxmax() 
print("\n MOST PROFITABLE CUSTOMER SEGMENT:", (best_segment))
```

## Export Results


```python
analysis_results = {
'total sales': total_sales,
'total_profit':total_profit,
'total_orders': total_orders,
'avg_profit_margin': avg_profit_margin,
'top_category': top_categories.index[0],
'best_region':best_region,
'best customer_segment': best_segment
}
#Convert to DataFrame and save
results_df= pd.DataFrame([analysis_results])
results_df.to_csv("eda_analysis_summary.csv', index=False")
print("\n Analysis summary saved to eda_analysis_summary.csv")
```
